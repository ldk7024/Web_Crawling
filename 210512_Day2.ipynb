## 날짜별로 영화데이터 수집

### 5월 1일부터 5월 12일까지 영화데이터 가져오기

```python
title_list =[]
score_list=[]
rank_list =[]
date_list =[]
```

```python
for i in range(20210501, 20210512):    
    url= "https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&tg=0&date=20210510"+str(i)
    res = req.get(url)
    soup = bs(res.text, 'lxml')
    title = soup.select('div.tit5 > a') # 영화명
    score = soup.select('td.point') # 평점
    
    for j in range(len(title)):
        score_list.append(point[j].text)
        title_list.append(title[j].text.strip())
        rank_list.append(j+1)
        date_list.append(i) 
```

```python
print(len(title_list))
print(len(score_list))
print(len(rank_list))
print(len(date_list))
```

```python
dic = {'rank' : rank_list, 'title' : title_list, 'score' : score_list, 'date' : date_list}
```

```python
movie_date = pd.DataFrame(dic)
```

```python
movie_date.set_index("rank", inplace = True)
```

```python
movie_date
```

```python
movie_date.to_csv('movie_date.csv', encoding ='euc-kr') # 파일명, 인코딩방식
```

### 진행사항을 볼 수 있는 함수

```python
from tqdm import tqdm_notebook
```

```python
title_list =[]
score_list=[]
rank_list =[]
date_list =[]



for i in tqdm_notebook (days):    
    url= "https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&tg=0&date="+i
    res = req.get(url)
    soup = bs(res.text, 'lxml')
    title = soup.select('div.tit5 > a') # 영화명
    score = soup.select('td.point') # 평점
    
    for j in range(len(title)):
        score_list.append(point[j].text)
        title_list.append(title[j].text.strip())
        rank_list.append(j+1)
        date_list.append(i) 
```

```python
print(len(title_list))
print(len(score_list))
print(len(rank_list))
print(len(date_list))
```

```python
dic = {'rank' : rank_list, 'title' : title_list, 'score' : score_list, 'date' : date_list}
```

```python
movie_date = pd.DataFrame(dic)
```

```python
movie_date.set_index("rank", inplace = True)
```

```python
movie_date
```
